```python
regex = '\#(?P<num>.+?)\s확진자(\\t)?\((?P<rsd>.+?)\)\s?(\\t)?\/\s(?P<dat>.+)감염경로\s:\s(?P<pth>.+)증상(발현일\s\:)?\s(?P<sym>.+)조치\s:\s(?P<atn>.+)'
info = [re.search(regex, t) for t in symplist]
```

```python
# regex2 = "감염경로\s:\s(?P<pth>.+)증상(발현일\s\:)?\s(?P<sym>.+)조치\s:\s(?P<atn>.+)"
# info2 = [re.search(regex2, t) for t in symplist]
```

```python
res=[(i.group('num'), i.group('rsd'), i.group('dat'), i.group('pth'), i.group('sym'), i.group('atn'))for i in info if i is not None]
gangnam=pd.DataFrame(res, columns=('num2','rsd','cnd','pth','sym','atn'))

today = str(dt.datetime.today().date()).replace('-','')[2:]
gangnam['dat'] = today
numlist = []
for x in range(len(gangnam['num2'])
    num = '108000'+gangnam['num2'][x]
    numlist.append(num)
    
gangnam['num2'][1]
```
