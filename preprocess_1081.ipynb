```python
from bs4 import BeautifulSoup
import datetime as dt
import pandas as pd
import numpy as np
import re
from tqdm import tqdm

dat = dt.datetime.today().date() # 오늘 날짜
```

### 1. Raw File 불러오기

```python
df = pd.read_csv("200906one_shot.csv")
gangdong = BeautifulSoup(df['html'][3], 'html.parser')
```

### 2. Regex Parsing 하기

```python

rows = gangdong.select('#divTable1 .acodList')

regex = 'tle\"\>\s강동구(?P<num>.+)\s확진자\((?P<rsd>.+)\)\s\/\s(?P<cnd>.+)\<.+\n.+'
regex += '\n.+it\"\>...?.?.?.?.?.?\((?P<age>..?.?.?.)대.?.?.?\,\s(?P<sex>..)\)\,.+로\:\s(?P<msg>.+)\<'
infos = [re.search(regex, str(r)) for r in rows]

#'148번 확진자(10대 여성 / 성내2동, 1개동 공동주택) 관련 자가격리자 총 4명(동거가족 4)입니다.' 와 같은 형태일 경우
regex2 = 'tle\"\>\s강동구(?P<num>.+)\s확진자\((?P<rsd>.+)\)\s\/\s(?P<cnd>.+)'
regex2 += '\<.+\n.+\n.+it\"\>...?.?.?.?.?.?\((?P<age>..)대\s(?P<sex>..).\/'

infos2 = [re.search(regex2, str(r)) for r in rows]
```

### 3. 저장할 DataFrame 미리 만들기

```python
pat_col = ['num', 'dat', 'pth', 'syy', 'syd', 'cnd', 'rsd', 'atn', 'dch', 'adt','age','sex']
rou_col = ['num', 'ord', 'dat', 'rgl', 'rgt', 'frm', 'exd', 'msk', 'ste', 'mob']
patient = pd.DataFrame(columns = pat_col)
route = pd.DataFrame(columns = rou_col)
```

### 4. 귀찮은 전처리해주기

```python
for i, info in tqdm(enumerate(infos)):
    
    # regex 처리가 안 되는 애들은 어차피 정보가 없으므로 None 값임.
    if info is not None:
        
        num = '1080' + '0' * (5-len(info.group('num'))) + info.group('num')
        syy = ''; syd = ''
        dch = ''; adt = '' # 정보가 없으면 이렇게 처리하자
        rsd, cnd, age, sex = info.group('rsd'), info.group('cnd'), info.group('age'), info.group('sex')
        
        #뒤 메시지가 중구난방이어서 ..
        msg = info.group('msg').split(', ')
        pth = msg[0]
        if len(msg)>1:
            if '이송병원' in msg[1]:
                msg[1] = msg[1].replace(' ','')
                index = msg[1].index(':') + 1
                atn = msg[1][index:]
        else:
            atn = ''
        
        new_pat = pd.DataFrame([(num, dat, pth, syy, syd, cnd, rsd, atn, dch, adt, age, sex)], columns = pat_col)
        patient = pd.concat([patient, new_pat])
        
        try: #table이 아예 없는 경우 존재
            new_route = pd.read_html(str(rows[i].select('table')[0]))[0]
            new_route['rgl'] = ''; new_route['msk'] = ''; new_route['ste'] = ''; new_route['rgt'] = ''
            
            new_route['exd'] = new_route.날짜+new_route.iloc[:,2]
#           
            new_route['num'] = num
            new_route['ord'] = range(1, len(new_route)+1)
            new_route['dat'] = dat
            new_route = new_route.drop(['날짜', '역학조사 결과'], axis = 1)
            new_route = new_route.drop(new_route.columns[1], axis=1)

            new_route.columns = ['frm', 'mob', 'rgl', 'msk', 'ste', 'rgt', 'exd', 'num', 'ord', 'dat']
            new_route = new_route.loc[:,rou_col]
            route = pd.concat([route, new_route])
        except:
            continue  
```
